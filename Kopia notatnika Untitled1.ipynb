```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df = pd.read_csv('salary_data - Salary_Data (3).csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price',axis='columns')
new_df
```

```python
price = df.price
price
```

```python
area_df = pd.read_csv("areas - Salary_Data (1).csv")
area_df.head(3)
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,price)
```

```python
reg.predict(area_df)
```

```python
p = reg.predict(area_df)
p
```

```python
area_df['prices']=p
area_df
```

```python
%matplotlib inline
plt.xlabel('area', fontsize=20)
plt.ylabel('price(US$)', fontsize=20)
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
